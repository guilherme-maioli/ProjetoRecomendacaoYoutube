```python
import youtube_dlc
import pandas as pd
import numpy as np
```

```python
queries = ["machine+learning", "data+science", "kanggle"]
```

```python
ydl = youtube_dlc.YoutubeDL({"ignoreerrors": True})
```

```python
resultados = []
for query in queries:
    r = ydl.extract_info("ytsearchdate10:{}".format(query), download=False)
    for entry in r['entries']:
        if entry is not None:
            entry['query'] = query
    resultados += r['entries']
```

```python
resultados = [e for e in resultados if e is not None]
```

```python
len(resultados)
```

```python
df = pd.DataFrame(resultados)

df['upload_date'] = pd.to_datetime(df["upload_date"])
df['tempo_desde_pub'] = (pd.to_datetime("2020-10-26") - df['upload_date']) / np.timedelta64(1, 'D')
```
